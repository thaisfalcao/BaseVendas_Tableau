# Verificando a existência de duplicatas

```python
df[df.duplicated()]
```

```python
df_drop = df.drop_duplicates().reset_index()

print(f'O arquivo contém {len(df.index) - len(df_drop.index)} linha(s) duplicata(s)')
```

# Verificando a existência de dados nulos

```python
df_drop.isnull().sum()
```

# Salvando a tabela após removação da linha repetida

```python
df_drop.to_excel('BaseVendas - sem duplicatas.xlsx')
```
